```python

#1. Load the dataset and handle any missing data by replacing null values with the mean
#value of the respective column. (Score 2)

import pandas as pd
import numpy as np

df = pd.read_csv('winequality-red.csv',sep=';')
#df

#missing data by mean
df.fillna(df.mean(), inplace=True)
#df
```

Select two features (e.g., alcohol and density) from the dataset and calculate the
covariance matrix using np.cov(X.T), where X is the feature matrix consisting of the
selected columns.(Score 2)

```python
X = df[['alcohol','density']].to_numpy()
#X

cv= np.cov(X.T)
cv
```

Perform eigen decomposition on the covariance matrix you computed in question 3.
Identify and interpret the results:Identify the top 2 eigenvalues of the covariance
matrix,Identify the corresponding eigenvectors.(Score 2)

```python
eigenvalues, eigenvectors = np.linalg.eig(cv)
print(eigenvalues)
print(eigenvectors)
# Sort them in descending order to easily find the top 2)
idx = eigenvalues.argsort()[::-1]
print(idx)
eigenvalues = eigenvalues[idx]
eigenvectors = eigenvectors[:, idx]
print(f"Eigen Values:\n {eigenvalues}")
print(f"Eigen Vectors:\n {eigenvectors}")
```
