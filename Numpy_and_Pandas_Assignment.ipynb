Create a dataframe with the following columns: name, age, and gender. The dataframe
should have 10 rows of data. (Score: 2)
1) Add a new column to the data frame created in question 1, called occupation.
The values for this column should be Programmer, Manager, and Analyst,
corresponding to the rows in the dataframe. (Score: 1)
2) Select the rows of the dataframe where the age is greater than or equal to 30.
(Score: 1)
3) Convert this dataframe to a CSV file and read that CSV file, and finally display
the contents. (Score: 1)

```python
import pandas as pd
#dict
data = {
    'name': ['Jishnu', 'Vishnu', 'Nandana', 'Nithin', 'Vipin', 'Vineetha', 'Asha', 'Sindhu', 'Veena', 'Manu'],
    'age': [25, 34, 22, 45, 29, 31, 28, 38, 40, 24],
    'gender': ['F', 'M', 'M', 'M', 'F', 'M', 'F', 'F', 'M', 'M']
}
df = pd.DataFrame(data)
print("Initial dataframe")
print(df)

occupations = ['Programmer', 'Manager', 'Analyst', 'Programmer', 'Manager',
               'Analyst', 'Programmer', 'Manager', 'Analyst', 'Programmer']
df['occupation'] = occupations
print("\n Dataframe with Occupation Column ")
print(df)

filtered_df = df[df['age'] >= 30]
print("\n Filtered Dataframe")
print(filtered_df)

csv_file_path = 'data_output.csv'
filtered_df.to_csv(csv_file_path)
print(f"\n Dataframe saved to {csv_file_path}")
```
