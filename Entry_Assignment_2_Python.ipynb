```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_excel('ABC Company.xlsx')
df.columns
```

Preprocessing:
● Correct the data in the "height" column by replacing it with random
numbers between 150 and 180. Ensure data consistency and integrity
before proceeding with analysis. (1 mark)

```python
df['Height'] = np.random.randint(150, 180,size=len(df))
df['Height']
```

Analysis Tasks:
● Determine the distribution of employees across each team and calculate
the percentage split relative to the total number of employees. (2 marks)

```python
#team_count = df['Team'].count
team_value_counts = df['Team'].value_counts()
print(f"Number of Emp:\n{team_value_counts}")
total_count = len(df)
print(f"Total Count:\n{total_count}")
#print(f"1:\n{team_count}")
team_per = team_value_counts/len(df)*100
print(f"Percentage:\n{team_per}")
team_summary = pd.DataFrame({
    'Employee Count': team_value_counts,
    'Percentage Split (%)': team_per.round(2)
})
print("--- Team Distribution Summary ---")
print(team_summary)
sns.barplot(x=team_value_counts.values, y=team_value_counts.index)
```

Segregate employees based on their positions within the company. (2
marks)

```python
count_position = df['Position'].value_counts()
print(f"Number of Emp:\n{count_position}")
sns.countplot(data=df, x='Position',order=df['Position'].value_counts().index)
plt.title('Position Distribution')
plt.show()
```

Identify the predominant age group among employees. (2 marks)

```python
#df['Age'].max()
#df['Age'].min()
df['Age'].value_counts()
sns.countplot(data=df, x='Age',order=df['Age'].value_counts().index)
plt.title('Age Distribution')
plt.show()
```

Discover which team and position have the highest salary expenditure. (2
marks)

```python
team_salary = df.groupby('Team')['Salary'].sum().sort_values(ascending=False)
print(f"Team Salary:\n{team_salary}")
position_salary = df.groupby('Position')['Salary'].sum().sort_values(ascending=False)
print(f"Position Salary:\n{position_salary}")
print("\n--- Top Expenditure Drivers ---")
print(f"Highest Spending Team: {team_salary.index[0]} - {team_salary.iloc[0]}")
print(f"Highest Spending Position: {position_salary.index[0]} - {position_salary.iloc[0]}")
```

Investigate if there's any correlation between age and salary, and
represent it visually. (2 marks)

```python
corr = df['Age'].corr(df['Salary'])
print(f"Correlation between Age and Salary: {corr}")
sns.scatterplot(data=df, x='Age', y='Salary', alpha=0.6, color='purple')
plt.title('Age vs Salary')
plt.show()
```
